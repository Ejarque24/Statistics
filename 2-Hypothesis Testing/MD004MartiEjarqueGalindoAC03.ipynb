El dataset adjunto incluye los datos finales del test y contiene tres columnas. ### 1. Exploración de los datos: análisis descriptivos de los datos, gráficación y conclusiones ¿siguen una distribución normal?(2pts)

```r
data <- read.csv(file='202311s20_b2b_ecommerce_sales_data.csv', header=TRUE, sep=',', dec='.')
```

```r
names(data)
dim(data)
```

Primeramente, podemos determinar que tenemos **3** variables para cada una de las **11786** compras realizadas en los tres meses seleccionados. 

```r
head(data)
summary(data)
str(data)
```

Gracias a la función **summary()**, podemos observar comportamientos interesantes de nuestros datos. Primeramente, los valores de venta son muy **dispersos**, donde nuestro mínimo son **0,38** (valor de venta, podemos suponer euros) y el máximo **9476,80**. Otra observación es la "distribución" de ventas con sistema de recomendación y sin. Podemos ver como la media no es exactamente 0.5, por lo que podemos, de antemano, intuir que **tenemos más ventas con sistema de recomendación que sin**. 

Antes de pasar a la creación del test, veamos como se comportan dichos datos. Primeramente, comprobamos que no haya identificadores de venta repetidos (para detectar posibles errores, o ventas hechas en ambos casos (0 y 1)):

```r
print(any(duplicated(data$References)))

print(table(data$Group))
```

Podemos comprobar como no hay ningún ID repetido, por lo que, a priori, cada **venta es independiente**. También se ha aprovechado para confirmar lo observado en primera instáncia: Tenemos mas ventas con sistema de recomendación que sin, creando un **desbalance** entre datos. Este hecho puede afectar a la hora de confeccionar e interpretar los resultados dados por el test de hipótesis en cuestión. 

Ahora, pasamos a representar la distribución de las ventas por clase, así como un boxplot para cada clase (con sistema de recomendación y sin).

```r
par(bg = "white")
hist(data$Order.Value[data$Group == 1], 
     main = "Histograma - Con Sistema de Recomendación",
     xlab = "Valor de la Venta",
     ylab = "Frecuencia",
     col = "lightblue",
     breaks = 75) # raíz cuadrada del número de instancias de clase 1 

hist(data$Order.Value[data$Group == 0], 
     main = "Histograma - Sin Sistema de Recomendación",
     xlab = "Valor de la Venta",
     ylab = "Frecuencia",
     col = "lightgreen",
     breaks = 80) # raíz cuadrada del número de instancias de clase 0 

boxplot(data$Order.Value ~ data$Group,  
        main = "Boxplot del Valor de las Ventas por Clase",
        xlab = "Clase",
        ylab = "Valor de la Venta",
        col = c("lightblue", "lightgreen"))
```

De ambos histogramas podemos apreciar como no sigue una distribución normal, sino una **exponencial**. En este caso, gran parte de las ventas están concentradas en valores de venta bajos comparado con todo el espectro de valores (entre 0 y 2000), en especial en rangos muy bajos (0-500). De hecho, en los boxplot podemos observar como hay una gran dispersión de valores de venta a partir de los 2000 aprox. 

Aun así, debemos concentrarnos en la distribución de nuestros datos. Las técnicas que usaremos para realizar el test de hipótesis pertinente asumen una distribución normal. No es nuestro caso, y por eso podemos estar tentados en concluir que debemos transformar nuestros datos o incluso usar otros métodos estadísticos (métodos no paramétricos). Pero, gracias al **teorema del límite central**, el cual podemos aplicar dado el número de observaciones adquirdos, podemos asumir que nuestros datos acaban adoptando una distribución normal, y por ello pueden ser usados en un test de hipotesis. [1]

A continuación, mostramos la demostración del teorema del límite central con los datos presentes en **Order.Value**

```r
generar_medias_muestrales <- function(tamano_muestra) {
  medias_muestrales <- replicate(1000, mean(sample(data$Order.Value, size = tamano_muestra, replace = TRUE)))
  return(medias_muestrales)
}

par(mfrow = c(4, 2), mar = c(4, 4, 2, 1), bg = "white")

tamanos_muestra <- c(10, 30, 50, 100)

for (i in 1:length(tamanos_muestra)) {
  medias_muestrales <- generar_medias_muestrales(tamanos_muestra[i])
  
  hist(medias_muestrales, 
       main = paste("Histograma (n =", tamanos_muestra[i], ")"),
       xlab = "Media Muestral",
       col = "skyblue",
       border = "black")
  
  curva_normal <- dnorm(seq(min(medias_muestrales), max(medias_muestrales), length.out = 100), 
                        mean = mean(data$Order.Value), 
                        sd = sd(data$Order.Value) / sqrt(tamanos_muestra[i]))
  
  lines(seq(min(medias_muestrales), max(medias_muestrales), length.out = 100), 
        curva_normal * length(medias_muestrales) * diff(hist(medias_muestrales)$breaks[1:2]), 
        col = "red", lwd = 2)
}
```

Podemos observar como, en diferentes agrupaciones, las medias de cada una de las agrupaciones, al final acaba conformando una distribución normal. Dicho código ha estado muy inspirado en la siguiente fuente [2]

### 2. Diseña un test de hipótesis: ¿cuál es nuestro objetivo? ¿se trataría de un test de 2 colas? ¿Qué estadístico debemos usar para este Test?(2pts)

Una vez comprobado los datos dados y sus requisitos para el desarrollo del test de hipótesis, procedemos a su diseño:

- **Objetivo:** Validar el uso del sistema de recomendación en el proceso de venta. Para ello, debemos comprobar si el valor medio de las ventas hechas con el sistema de recomendación supera a las hechas sin él, para concluir si dicho sistema favorece a la empresa a ganar más dinero o, por lo contrario, no representa una diferencia significativa.

- **Estadístico**: Por lo que concierne al estadístico, hay varios factores a tener en cuenta. En primera instáncia, podemos pensar en usar una distribución Z, ya que tenemos un número de observaciones apto para ello. Aun así, hay un valor clave el cual desconocemos, la desviación estándar de la **población**, dato necesario para el cálculo del valor z. Para ello, aunque tengamos muchas observaciones (>30), optaremos por una **T de Student**. Al final, si superamos las 30 observaciones, el estadístico se comporta como una normal. 

![z-stadistic](<https://miro.medium.com/v2/resize:fit:287/1*b7iZyQyP8SJ-W51x_L5ekg.png>)



- **Test**: Dicho test de Hipótesis se ha decidido de hacerlo de una cola, en la que:

      1. Hipótesis nula:
     + $H_{0}: µ_{con sistema} = µ_{sin sistema}$ (el sistema de recomendación no afecta en el valor de las ventas)
 
      2. Hipótesis alternativa: 
    + $H_{1}: µ_{con sistema} > µ_{sin sistema}$ (el sistema de recomendación ayuda a generar más valor de venta que sin él).

 

### 3. Validación y ejecución del test de hipótesis (2pts) 

Tal y como hemos mencionado en el apartado anterior, usaremos una **T de Student**. Para respetar el test de una cola, usaremos como alternative "greater". Para ello, el valor X debe ser las ventas con sistema, y las Y, los sin.

Siguiendo la documentación encontrada [3], la hipótesis alternativa se define como $H_{1}: µ_{x} > µ_{y}$, siendo X el primer parámetro de la función **t.test()**, y la Y, el segundo. 




```r
resultado_test <- t.test(
        data$Order.Value[data$Group == 1], 
        data$Order.Value[data$Group == 0],
        alternative = "greater")

print(resultado_test)
```

Con los resultados obtenidos, podemos observar como el p-value es **menor al 0,05** (por defecto la alpha en la función es del 95%), por lo que **la hipótesis nula queda descartada**. De hecho, podemos ver como en el estadístico, el valor t está casi a **3 desviaciones estándar** de distancia por encima de la media hipotética bajo la hipótesis nula. 

Por lo que, a priori, la media de ventas es superior cuando el sistema de recomendación está activado. De hecho, podemos observar como las media de x (con sistema) es superior a la de y (sin sistema).


```r
efecto_real <- 150 
nivel_significancia <- 0.05
potencia_deseada <- 0.95

muestras_aleatorias <- sample(data$Order.Value[data$Group == 1], length(data$Order.Value[data$Group == 0]))

s_p <- sd(muestras_aleatorias - data$Order.Value[data$Group == 0])

tamaño_muestra <- power.t.test(delta = efecto_real, 
                                sd = s_p, 
                                sig.level = nivel_significancia, 
                                power = potencia_deseada, 
                                type = "one.sample",
                                alternative = "one.sided")$n

tamaño_muestra <- ceiling(tamaño_muestra)

cat("Tamaño de muestra necesario:", tamaño_muestra)

```

Por último, tal y como se hace en el ejercicio de ejemplo, calcularemos el tamaño muestral necesario para igualar ambos errores (Tipo I y II). Para ello, contamos con un pequeño inconveniente: los datos están desbalanceados. Para resolver dicha problemática, se ha decidido "igualar" los datos, seleccionando tantos datos del grupo 1 como el total en el grupo 0, de manera aleatoria.

Después de varias ejecuciones, hemos observado como los valores rondan entre los **7400 y 7800**, por lo que concluimos que tenemos una muestra lo suficientemente grande como para que **ambos errores se igualen**.

Si basamos nuestro razonamiento puramente en el output del test de hipótesis, a priori podemos concluir que el sistema de recomendación cumple con su objetivo: aumentar las ventas para nuestra empresa. Ahora bien, considero que, debido a la naturaleza de los datos que tenemos, dichos resultados **no son suficientes para recomendar su uso**.

Hay información muy importante con la cual no contamos y que puede influenciar significativamente los resultados de dicho test. Primeramente, no sabemos en qué **sector** el ecommerce actua. Dicha información tiene un peso importante para saber interpretar los valores adjuntos en el fichero csv. Dicho esto, cierto es que podemos intuir que se trata de un ecommerce relativamente **grande o polifacético**, ofreciendo productos en un rango de valores muy amplio (0.38 - 9476.80). Aun así, el sector es determinante, ya que dependiendo de este, hay **épocas del año** más o menos fructíferas para la empresa, o incluso la **oferta y demanda** pueden verse afectadas.

Este último punto va relacionado con el lapso temporal de dichos datos, 3 meses. No sabemos en qué meses se han **recolectado los datos**, lo que implica que, sumado al sector, sea de vital importancia para entenderlos. Si se trata de un producto que se consume en su mayoría durante la época de Navidad, y dichos datos están recolectados en 3 meses consecutivos de verano/primavera, los valores no serán representativos de la demanda de dicho producto en la época correspondiente. 

Es por eso que, aunque tengamos suficiente muestra de datos para asegurar un buen test de hipótesis, como mínimo exploraría hacer dicho test con datos de un **año fiscal completo**, o en su defecto, teniendo conocimiento de qué meses estamos tratando, para actuar en consecuencia. Hablando de muestras, otro aspecto a tener en cuenta es el **desbalance** de datos. El hecho de tener más datos de cuando se usa el sistema de recomendación en comparación de cuando no, puede introducir un **sesgo en los datos**, alterando la interpretación del test de hipótesis. Tener un equilibrio de dichos datos sería crucial para estar más seguro a la hora de recomendar la herramienta.

En definitiva, aunque el test de hipótesis de luz verde al uso del sistema de recomendación, la base con la cual se ha hecho dicho test no es lo suficientemente robusta (bajo mi punto de vista) como para dar validez a dicho resultado.
